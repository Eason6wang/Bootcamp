```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('document.jpeg')
plt.figure(figsize=(20,20))
plt.imshow(image,cmap='gray')
```

```python
from wand.image import Image
from PIL import Image
import pyocr
import pyocr.builders
import io
'''
https://gitlab.gnome.org/World/OpenPaperwork/pyocr
brew install tesseract
pip install pyocr
pip install wand
pip install pillow"
brew install freetype imagemagick
'''
```

```python
tool = pyocr.get_available_tools()[0]
lang = tool.get_available_languages()[0] #english
```

```python
### Detection of words
image = cv2.imread('document.jpeg')
word_boxes = tool.image_to_string(
    Image.open('document.jpeg'),
    lang=lang,
    builder=pyocr.builders.WordBoxBuilder()
)
for box in word_boxes:
    cv2.rectangle(image, box.position[0], box.position[1], color=(0,0,255), thickness=2)
    cv2.putText(image, box.content, box.position[0],0,fontScale=1,color=(255,0,0),thickness=2)
plt.figure(figsize=(20,20))
plt.imshow(image,cmap='gray')
```

```python
### Detection of lines
image = cv2.imread('document.jpeg')
line_and_word_boxes = tool.image_to_string(
    Image.open('document.jpeg'), lang=lang,
    builder=pyocr.builders.LineBoxBuilder()
)
for box in line_and_word_boxes:
    cv2.rectangle(image, box.position[0], box.position[1], color=(0,0,255), thickness=2)
    cv2.putText(image, box.content, box.position[0],0,fontScale=1,color=(255,0,0),thickness=2)
plt.figure(figsize=(30,30))
plt.imshow(image,cmap='gray')
```
